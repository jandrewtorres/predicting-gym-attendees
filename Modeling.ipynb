# Modeling

## Import and read_csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/data/jandrewtorres/data.csv')
```

## Split/Create test and train data

```python
column_names = ['day_of_week', 'is_weekend',
       'is_holiday', 'temperature', 'is_start_of_semester',
       'is_during_semester', 'month', 'hour']
X = df[column_names]
y = df['number_people']
y = np.array(y, dtype='float64')
X = X.astype('float64')
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

## Linear Regression Model

### Lets look at a basic Linear Regression set-up and see how it does

```python
lin_model = LinearRegression()
lin_model.fit(Xtrain, ytrain)
lin_model.score(Xtest, ytest)
```

Better than guessing, but not what I want. Lets try another option:
    Add a PolynomialFeatures transformer and see how that does
    
### Plot the score of varying degrees of PolynomialFeatures transform

```python
x_ax = [i for i in range(5)]
scores = []
for i in range(5):
    poly_lin_model = make_pipeline(PolynomialFeatures(i), LinearRegression())
    poly_lin_model.fit(Xtrain, ytrain)
    ytrainpred = poly_lin_model.predict(Xtrain)
    ytestpred = poly_lin_model.predict(Xtest)
    scores.append(poly_lin_model.score(Xtest, ytest))
```

```python
plt.bar(x_ax, scores)
plt.title('Polynomial Features vs Score')
plt.xlabel('Num of Polynomial')
plt.ylabel('Score')
plt.xticks(x_ax)
plt.show()
```

We can see that PolynomialFeatures(4) is most effective
### Results

```python
print('Score: ',poly_lin_model.score(Xtest, ytest))
print("r squared test (training): ", r2_score(ytrain, ytrainpred))
print("r squared test (prediction): ", r2_score(ytest, ytestpred))
```

## Random Forest Regression

```python
model = RandomForestRegressor()
model.fit(Xtrain, ytrain)
model.score(Xtest, ytest)
```

Thats more like it!

A grid search is

```python
model = RandomForestRegressor()

param_grid = { 
   'n_estimators': [100, 200, 400],
}

CV_rfc = GridSearchCV(estimator=model, param_grid=param_grid, cv= 2)
CV_rfc.fit(X, y)
print(CV_rfc.best_params_)
```

### Use these hyperparameters in RandomForestRegressor

```python
model = RandomForestRegressor(max_features='log2', n_estimators=400)
model.fit(Xtrain, ytrain)
ytestpred = model.predict(Xtest)
ytrainpred = model.predict(Xtrain)
```

```python
print('Score: ', model.score(Xtest, ytest))
print("r squared test(training): ", r2_score(ytrain, ytrainpred))
print("r squared test(prediction): ", r2_score(ytest, ytestpred))
```

### Plot Actual vs. Predicted

```python
x_axis = np.arange(0, 200, 1)
plt.figure(figsize=(16,4))
plt.plot(x_axis, ytest[:200], label='actual')
plt.plot(x_axis, ytestpred[:200], label='predicted')
plt.title('Actual number of people vs. Predicted number of people')
plt.xlabel('Arbitrary index number')
plt.ylabel('Number of People')
plt.legend()
plt.show()
```
